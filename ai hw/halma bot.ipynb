```python
def mandis(p1,p2):
    return ((abs(p1[0]-p2[0])+abs(p1[1]-p2[1])))
```

```python
def minmandis(p1,p2):
    return (max(abs(p1[0]-p2[0]),abs(p1[1]-p2[1])))
```

```python
def euc(p1,p2):
    return math.sqrt((p1[0]-p2[0])**2 + (p1[1]-p2[1])**2 )
```

```python
def utildist(p1,p2):
    m=max(p2[0],p2[1])
    return (m+1)*minmandis(p1,p2)
```

```python
def nxt(board,m):
    boards=copy.deepcopy(board)
    m[1]=m[1].split(',')
    m[2]=m[2].split(',')
    #print(m[1],m[2])
    boards[int(m[2][1])][int(m[2][0])]=boards[int(m[1][1])][int(m[1][0])]
    boards[int(m[1][1])][int(m[1][0])]='.'
    #print("orig",board)
    #print("modi",boards)
    return boards
    
```
