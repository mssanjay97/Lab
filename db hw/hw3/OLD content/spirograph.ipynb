```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
R=0.0005;r=0.0001;a=0.0003
#x_c=0;y_c=0
x_c=34.020936;y_c=-118.285566
```

```python
xpts=[]
ypts=[]
rev=10
t=0
while (t<rev*math.pi):
    x=x_c+(R+r)*math.cos((r/R)*t) - a*math.cos((1+r/R)*t)
    y=y_c+(R+r)*math.sin((r/R)*t) - a*math.sin((1+r/R)*t)
    xpts.append(x)
    ypts.append(y)
    t+=0.01
```

```python
len(xpts)
```

```python
plt.plot(xpts,ypts)
plt.show()
```
